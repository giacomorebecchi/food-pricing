```python
from src.data.storage import get_S3_fs
from src.features.csv_to_parquet import (
    load_csv,
    ITEM_DTYPES,
    ITEM_INPUT_PATH,
    ITEM_SUFFIX,
)
import dask.dataframe as dd
from dotenv import load_dotenv

load_dotenv()
```

```python
S3 = get_S3_fs()
csv_paths = S3.glob("s3://" + ITEM_INPUT_PATH)
delayed_dfs = (load_csv(fp, ITEM_SUFFIX, ITEM_DTYPES) for fp in csv_paths)
ddf = dd.from_delayed(
    delayed_dfs,
    meta={
        **ITEM_DTYPES,
        "filePath": str,
        "city": str,
        "zone": str,
        "store": str,
        "menuRow": int,
    },
).repartition(partition_size="10MB")
```
