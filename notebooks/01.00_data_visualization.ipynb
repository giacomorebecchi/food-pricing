```python
import dask.dataframe as dd
from src.data.storage import dd_read_parquet
from src.data.config import (
    COORDINATES_TABLE,
    FULL_TABLE,
    IMAGES_TABLE,
    ITEMS_TABLE,
    Table,
)

REMOTE = True
```

```python
def show_table(table: Table):
    path = table.remote_path if REMOTE else table.local_path
    ddf = dd_read_parquet(
        path=path,
        remote=REMOTE,
    )
    return ddf.head()
```

```python
show_table(COORDINATES_TABLE)
```

```python
show_table(IMAGES_TABLE)
```

```python
show_table(ITEMS_TABLE)
```

```python
show_table(FULL_TABLE)
```

```python
def check_table_length():
    path_full = FULL_TABLE.remote_path if REMOTE else FULL_TABLE.local_path
    path_items = ITEMS_TABLE.remote_path if REMOTE else ITEMS_TABLE.local_path
    assert (
        len(dd_read_parquet(path_full, REMOTE)) == len(dd_read_parquet(path_items, REMOTE))
        )

check_table_length()
```
